### Analyze the data


```python

megaline_calls['month'] = megaline_calls['call_date'].dt.month

#The number of calls made and minutes used per month:

total_user_calls = megaline_calls.groupby(['user_id', 'month'])['duration'].sum().reset_index()

total_user_calls.rename(columns={'duration': 'minutes_used'}, inplace=True)

total_user_calls1 = megaline_calls.groupby(['user_id', 'month'])['duration'].count().reset_index()

total_user_calls['calls_made'] = total_user_calls1['duration']

print(total_user_calls)

megaline_internet['month'] = megaline_internet['session_date'].dt.month

#The volume of data per month:

total_internet_use = megaline_internet.groupby(['user_id', 'month'])['mb_used'].sum().reset_index()

print(total_internet_use.sort_values('mb_used'))

total_internet_use.rename(columns={'mb_used': 'Gb_used'}, inplace=True) # I renamed the column name 

total_internet_use['Gb_used'] = total_internet_use['Gb_used'].apply(lambda x: int(np.ceil(x / 1024.0))) # Rounded up
      
print(total_internet_use.sort_values('Gb_used'))        

megaline_messages['month'] = megaline_messages['message_date'].dt.month

#The number of text messages sent per month:

total_user_messages = megaline_messages.pivot_table(index = ['user_id', 'month'], values = 'id', aggfunc='count')

total_user_messages = total_user_messages.rename(columns={'id': 'messages_per_user'})

print(total_user_messages)

total1= pd.merge(left = total_user_calls, 
                 right = total_internet_use, how='outer',on=['user_id', 'month']).fillna(0)

users_total=pd.merge(left = total1, 
                right = total_user_messages, how='outer',on=['user_id', 'month']).fillna(0)

print(len(users_total))

users_total
```

<div class="alert alert-block alert-success">
<b>Reviewer's comment</b>

In order to calculate the average profit per user I rounded up the monthly mb internet usage and converted it to Gb according to Megaline policy, based on this company logic I treated new users same as existed for monthly plan charge. 

```python

User_costs = pd.merge(users_total, megaline_users[['tariff','user_id','city']], how='outer',on='user_id').fillna(0)

print(User_costs)

def cost(row):
    tariff=row['tariff']
    calls=row['minutes_used']
    text=row['messages_per_user']
    Gb=row['Gb_used']
    calls_cost=0
    text_cost=0
    Gb_cost=0
    total_cost=0
    if tariff=='surf':
        user_cost=20
    else:
        user_cost=70
    if tariff=='surf' and calls-500 > 0:
        calls_cost = (calls-500) * 0.03
    elif tariff=='ultimate' and calls-3000 > 0:
        calls_cost = (calls-3000) * 0.01
    if tariff=='surf' and text-50 > 0:
        text_cost = (text-50) * 0.03
    elif tariff=='ultimate' and text-1000 > 0:
        text_cost = (text-1000) * 0.01
    if tariff=='surf' and Gb-15 > 0:
        Gb_cost = (Gb-15) * 10
    elif tariff=='ultimate' and Gb-30 > 0:
        Gb_cost = (Gb-30) * 7
    total_cost = user_cost + calls_cost + text_cost + Gb_cost
    return total_cost

User_costs['total_cost'] = User_costs.apply(cost, axis=1)

User_costs.sort_values('total_cost')           
```

### Test the hypotheses:
The average profit from users of Ultimate and Surf calling plans differs.
The average profit from users in NY-NJ area is different from that of the
users from other regions.

```python
# random sample:
sample_surf=surf_plan.sample(n=len(ultimate_plan), random_state=1)

print(sample_surf['total_cost'].var()) 

print(ultimate_plan['total_cost'].var())

# variance are different 

alpha = 0.05 # critical statistical significance
equal_var = False
results = st.ttest_ind(
        sample_surf['total_cost'], 
        ultimate_plan['total_cost'])

print('p-value: ', results.pvalue)

if (results.pvalue < alpha):
        print("We reject the null hypothesis")
else:
        print("We can't reject the null hypothesis") 
        
#
```

In order to test the hypothes if the average profit from users of Ultimate and Surf calling plans differs I apply the stats.ttest_ind method.
The null hypothesis is that the average profits are equal.
In order to make groups equal I took random simple from the larger surf_plan group. Variance calculation of two groups showed that two groups have different variance, inorder to get the most accurate results equal_var param was set to False.
According to analysis average profit from users of Ultimate and Surf calling plans are different.
